```python
def get_country(address):
    
    found_matches = []

    a = pycountry.countries.get(name = address)
    b = pycountry.countries.get(alpha_3 = address)
    c = pycountry.countries.get(alpha_2 = address)
    if address == "uk":
        return "United Kingdom"
    elif a != None:
        found_matches.append((a.name))
    elif b != None:
        found_matches.append((b.name)) 
    elif c != None:
        found_matches.append((c.name)) 
    
    if found_matches != []:
        return found_matches
    else:
        return address
    
```
